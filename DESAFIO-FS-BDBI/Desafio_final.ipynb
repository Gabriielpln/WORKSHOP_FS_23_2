## Agrupamento de Dados e Plotagem de Gráficos

```python
df.head()
```

> - Como não temos o nome dos funcionários ou id para fazer contagem, implementei a aseguinte solução:

```python
# Agrupando Pelos Valores Únicos no Campo de Department.
count_setores = df["Department"].value_counts()
```

```python
count_setores
```

```python
# Plot
count_setores.plot(kind='bar', color="orange")
plt.title('Quantidade de Funcionários por Setor')
plt.xlabel('Setor')
plt.ylabel('Quantidade de Funcionários')
plt.show()
```

> Aqui, observamos a distribuição de funcionários por setor. O setor de pesquisa e desenvolvimento é o maior, com um total de 961 funcionários, seguido pelo setor de vendas, que conta com 446 funcionários. Em terceiro lugar, encontramos o setor de Recursos Humanos, com um total de 63 funcionários.

```python
# Agrupando Nível de Escolaridade
df2 = df.groupby("Education")[["Department"]].count()
```

```python
# Agrupado
df2
```

```python
df2.plot(kind='line', color="black")
plt.title('Qtd Funcionários por Nível de Escolaridade')
plt.xlabel('Nível Escolaridade')
plt.ylabel('Quantidade de Funcionários')
plt.show()
```

> Temos 5 níveis de escolaridade, sendo a categoria 5 a menos numerosa, com 48 funcionários, e a categoria 3 a mais populosa, contando com 572 funcionários.

```python
# Agrupando Atrito na Empresa
df3 = df.groupby("Attrition")[["Department"]].count()
```

```python
# Agrupado
df3
```

```python
cores = ["c", "green"]
df3.plot(kind = "pie", subplots = True, colors=cores)
```

> A análise revela que 237 funcionários têm algum tipo de atrito com a empresa, enquanto 1233 funcionários não apresentam atrito com a empresa.
